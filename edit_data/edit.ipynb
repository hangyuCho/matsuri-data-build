```python
import pandas as pd
import numpy as np
import json
import os
import re

from gmaps import Geocoding
```

```python
api = Geocoding(api_key='###')
```

```python
#pd.read_json('../data/page01.json')
dfs = []
for path in os.listdir('../data'):
    dfs.append(pd.read_json("../data/%s" % path))

df = pd.concat(dfs)
```

```python
df_distinct = df.copy().drop_duplicates()
df_distinct
```

```python
df_date = df_distinct.copy()
df_date
```

```python
df_date["開催日2"] = df_date["開催日"].map(lambda v:v.replace('・', '～')).map(lambda v:v.replace('～202','202'))
df_date["開始日"] = df_date["開催日2"].map(lambda v:v.split(')')[0] + ')')
df_date["終了日"] = df_date["開催日2"].map(lambda v:v.split(')')[1][1:] if len(v.split(')～')) > 1 else '').map(lambda v:v + ')' if v else v)
df_date["終了日"] = df_date["開始日"] + df_date["終了日"]
df_date["終了日"] = df_date["終了日"].map(lambda v: v if len(v) <= 14 else (v.split('月')[0] + '月' + v.split(")")[1] + ')' if len(v) <= 19 else v.split('年')[0] + '年' + v.split(")")[1] + ')'))
df_date["備考"] = df_date["開催日2"].map(lambda v:")".join(v.split(')')[1:]) if len(v.split(')～')) == 14 and len(v.split(')')) > 1 else ")".join(v.split(')')[2:]))
df_date
```

```python
df_date["座標"] = df_date["住所"].map(lambda v: api.geocode(v))
```
